```python
from transformers import pipeline


class TransClass:

        def __init__(self,text, model):
            self.text = text
            self.model = model

            if   model == "SENT":
                                   self.SENTIMENT_ANALYSIS()
            elif model == "ZERO":
                                   self.ZERO_SHOT_CLASSIFICATION()
            elif model == "TEXTG":
                                   self.TEXT_GENERATION()
            elif model == "MASKF":
                                   self.MASK_FILLING()
            elif model == "NER":
                                   self.NER()
            elif model == "Q&A":
                                   self.QANDA()

            elif model == "SUM":
                                      self.SUMMARIZATION()

            else :

                     print('Error')




        def SENTIMENT_ANALYSIS(self):

                 classifier = pipeline("sentiment-analysis")
                 output = classifier(self.text)
                 print(output)

        def ZERO_SHOT_CLASSIFICATION(self):

                classifier = pipeline("zero-shot-classification")
                output = classifier(self.text , candidate_labels=["education", "politics", "business" , "personal"],)
                print(output)

        def TEXT_GENERATION(self):

                generator = pipeline("text-generation")
                output = generator(self.text , max_length=100 , num_return_sequences = 5)
                print(output)

        def MASK_FILLING(self):
                unmasker = pipeline("fill-mask")
                output = unmasker(self.text , top_k = 5)
                print(output)


        def NER(self):
                ner = pipeline("ner" , grouped_entities=True)
                output = ner(self.text)


        def QANDA(self):
            question_answerer = pipeline("question-answering")
            output = question_answerer(
                             question=self.text,
                             context="His name is Senura Ruvinda Fernando whos good looking",
        )
            print(output)


        def SUMMARIZATION(self):
            summa = pipeline("summarization")
            output = summa(self.text)
            print(output)
```

```python
SENTIMENT_ANALYSIS = TransClass("I want to be a Millionare" , "SENT")
```

```python
TEXT_GENERATION = TransClass("I want to be a Millionare" , "TEXTG")
```
